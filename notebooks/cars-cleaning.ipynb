```python
## Importing Important Libraries!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Loadind Dataset!
df = pd.read_csv("datasets/cars_uncleaned.csv",encoding="latin1")
```

```python
df.shape # 1218 total records!
```

## Data Cleaning

```python
# making a copy dataset!
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
## Checking the data set!
df_copy.info()
```

```python
df_copy.isnull().sum()
```

```python
# removing NaN's as there are very less
df_copy = df_copy.dropna()
```

```python
df_copy.shape   ## Dropped 8 rows!
```

```python
## Checking for duplicates!
df_copy[df_copy.duplicated()].shape   ## 4 duplicates
```

```python
# Dropping duplicates
df_copy=df_copy.drop_duplicates()
```

```python
df_copy[df_copy.duplicated()].shape  ## Done
```

```python
## Now i am converting the columns to numerical in which it is feasable!
df_copy['CC/Battery Capacity'].unique()
```

```python
### For non ev Vehicles its in cc and for ev its in kwh!
# so first splitting it wrt spaces and including only numeric part i.e str[0]
df_copy['CC/Battery Capacity']=df_copy['CC/Battery Capacity'].str.split(" ").str[0]
df_copy['CC/Battery Capacity'].unique()
```

```python
## replacing ',' with empty ''..
df_copy['CC/Battery Capacity']=df_copy['CC/Battery Capacity'].str.replace(",","")
df_copy['CC/Battery Capacity'].unique()
```

```python
## there is row with '~' removing this sign!
df_copy['CC/Battery Capacity']=df_copy['CC/Battery Capacity'].str.replace("~","")
df_copy['CC/Battery Capacity']=df_copy['CC/Battery Capacity'].str.replace("HYBRID(","")
```

```python
## Seeing how many columns are left which are not numeric!
df_copy['CC/Battery Capacity'].unique()
```

```python
df_copy[~df_copy['CC/Battery Capacity'].str.isnumeric()]
```

```python
df_copy = df_copy.drop([994,995,1031,1032])
df_copy['CC/Battery Capacity']=df_copy['CC/Battery Capacity'].astype(float)
```

```python
print(df_copy.shape)
df_copy.info()    ## there are 1202 records left!
```

```python
## Now converting Horsepower Column!
df_copy['HorsePower'].unique()
```

```python
## there are some records with range of hp but i will only consider lower range for ease of cleaning
df_copy['HorsePower']=df_copy['HorsePower'].str.split(" ").str[0]
df_copy['HorsePower'].unique()
```

```python
## replacing the records with "~"
df_copy['HorsePower']=df_copy['HorsePower'].str.replace("~","")
df_copy['HorsePower'].unique()
```

```python
df_copy[~df_copy['HorsePower'].str.isnumeric()]
```

```python
## removing the ","
df_copy['HorsePower']=df_copy['HorsePower'].str.replace(",","")
df_copy[~df_copy['HorsePower'].str.isnumeric()]
```

```python
# also there is one row the non numerics hp with the median of numerics hp
numeric_hp = df_copy[df_copy["HorsePower"].str.isnumeric()]
numeric_hp['HorsePower'] = numeric_hp['HorsePower'].astype(int)
numeric_hp.info()
```

```python
## replacing non numeric column with median!
df_copy['HorsePower'] = df_copy['HorsePower'].replace(to_replace=r'.*Up.*', value=np.nan, regex=True)
```

```python
numeric_hp['HorsePower'].median()
```

```python
df_copy['HorsePower']=df_copy['HorsePower'].fillna('250.0')
```

```python
df_copy['HorsePower']=df_copy['HorsePower'].str.split("-").str[0]
```

```python
df_copy['HorsePower'] = df_copy['HorsePower'].astype(float)
```

```python
df_copy.info()
```

```python
## Now converting Total speed!
df_copy['Total Speed'].unique() 
```

```python
df_copy['Total Speed']=df_copy['Total Speed'].str.split(" ").str[0]
df_copy['Total Speed']=df_copy['Total Speed'].astype(float)
df_copy.info()
```

```python
## now converting 0-100 time!
df_copy['Performance(0 - 100 )KM/H'].unique()
```

```python
df_copy['Performance(0 - 100 )KM/H']=df_copy['Performance(0 - 100 )KM/H'].str.split(" ").str[0]
df_copy['Performance(0 - 100 )KM/H'].unique()
```

```python
df_copy[df_copy['Performance(0 - 100 )KM/H'] == ""]
```

```python
df_copy = df_copy.drop([91])
df_copy['Performance(0 - 100 )KM/H']=df_copy['Performance(0 - 100 )KM/H'].astype(float)
```

```python
df_copy.info()
```

```python
## Converting Car Prices!
df_copy['Cars Prices'].unique()
```

```python
df_copy['Cars Prices']=df_copy['Cars Prices'].str.split(" ").str[0]
df_copy['Cars Prices']=df_copy['Cars Prices'].str.replace(",","")
df_copy['Cars Prices']=df_copy['Cars Prices'].str.replace("$","")
df_copy['Cars Prices']=df_copy['Cars Prices'].str.split("-").str[0]
df_copy['Cars Prices']=df_copy['Cars Prices'].str.split("\x80").str[0]
df_copy['Cars Prices'].unique()
```

```python
df_copy[df_copy['Cars Prices']==""]
df_copy=df_copy.drop([1216])
df_copy['Cars Prices']=df_copy['Cars Prices'].astype(float)
```

```python
df_copy.info()
```

```python
# converting the seats
df_copy['Seats'].unique()
lst=['\x915','\x967','\x966','\x969','\x9612']
for item in lst:
    df_copy['Seats']=df_copy['Seats'].str.split(item).str[0]
df_copy['Seats']=df_copy['Seats'].str.replace("2+2","4")
df_copy['Seats'].unique()
```

```python
df_copy['Seats']=df_copy['Seats'].str.split("\x968").str[0]
df_copy['Seats']=df_copy['Seats'].str.split("\x9615").str[0]
df_copy['Seats']=df_copy['Seats'].astype(int)
```

```python
df_copy.info()
```

```python
## now only 1 column is left! converting that also
df_copy['Torque'].unique()
df_copy['Torque']=df_copy['Torque'].str.split(" ").str[0]
df_copy['Torque']=df_copy['Torque'].str.split("-").str[0]
df_copy['Torque']=df_copy['Torque'].str.replace(",","")
df_copy['Torque']=df_copy['Torque'].str.split("+").str[0]
df_copy['Torque'].unique()
```

```python
df_copy['Torque']=df_copy['Torque'].astype(float)
df_copy.info()
```

### Converted All The Required Columns to numeric type!

```python
## saving the dataset!
df_copy.to_csv("Datasets/cars_cleaned.csv")
```
